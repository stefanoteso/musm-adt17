```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.utils import shuffle
from sklearn import preprocessing, cross_validation, neighbors
import pandas as pd
df = pd.read_csv("/Users/bereket/Documents/Social/setmargin/musm-adt17/result_data/NORMALIZED_synthetic_loss_20_5_uniform.csv", names = ["Utility loss"])
df0 = pd.read_csv("/Users/bereket/Documents/Social/setmargin/musm-adt17/result_data/synthetic_TIME_20_5_uniform.csv", names = ["Cumulative time"])
r1 = pd.concat([df, df0], axis=1)
r1["Iteration "] = df.index + 1
r1[:5]
```

## Synthetic Experiment with (domain_sizes = 4, 20 Users, 1-2-5 Clusters, 100 iterations)

# 20 users, 1cluster, 100 iterations, Normal

```python
s = r1.ix[:,0]
t = r1.ix[:,1]
u = r1.ix[:,2]
Utility_loss = np.squeeze(np.asarray(s))
Cumulative_time = np.squeeze(np.asarray(t))
Iteration = np.squeeze(np.asarray(u))
#Ranking_loss
#Cumulative_time
#Iteration
```

# Plot Normalized Utility_loss (20 users, 1cluster, 100 iterations, Normal)

```python
plt.plot( Iteration, Utility_loss, color ='red')
plt.xlabel('Number of iternations')
plt.ylabel('Utility loss')
plt.plot([],[], color = 'red', label = 'independent', linewidth = 5)
plt.savefig('/Users/bereket/Documents/Social/setmargin/musm-adt17/Plots/Loss_synthetic_20_1_normal.png', transparent = True, bbox_inches = 'tight')
plt.legend()
plt.show()
plt.draw()

```

# Plot Cumulative time (20 users, 1cluster, 100 iterations, Normal)

```python
plt.plot( Iteration, Cumulative_time, color ='green')
plt.xlabel('Number of iternations')
plt.ylabel('Cumulative_time(s)')
plt.plot([],[], color = 'green', label = 'independent', linewidth = 5)
plt.savefig('/Users/bereket/Documents/Social/setmargin/musm-adt17/Plots/time_synthetic_20_1_normal.png', transparent = True, bbox_inches = 'tight')
plt.legend()
plt.show()
plt.draw()

```

# Plot(Loss and Time Correlation) 20 users, 1cluster, 100 iterations, Normal

```python
plt.plot(Iteration,Utility_loss, linewidth = 5, label = "Utility_loss")
plt.plot(Iteration,Cumulative_time, linewidth = 5, label = "Cumulative_time")
plt.xlabel('Number of iternations')
plt.ylabel('Loss and Time Correlation')

plt.legend()
plt.savefig('/Users/bereket/Documents/Social/setmargin/musm-adt17/Plots/Loss_+TIME_synthetic_20_1_normal_1.png', transparent = True, bbox_inches = 'tight')
plt.show()
plt.draw()
```
